```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('SsoObservation_-4284702096_-4283326086.csv')
epoch = data['epoch_utc']
epoch = epoch.sort_values(ascending=True)
dt = pd.Series(epoch.diff(), name='$\\bf{dt}$ in seconds')
```

```python
sns.set(rc={'figure.figsize':(12,6)})
ax = sns.distplot(dt, kde=False)
ax.set(yscale="log")
ax.set(ylabel="Number of observations")
```
